```python
driver=webdriver.Chrome()
driver.get('https://www.flipkart.com/')
driver.find_element(By.XPATH,'//input[@class="Pke_EE"]').send_keys('Singlasses')
driver.find_element(By.XPATH,'//input[@class="Pke_EE"]').send_keys(Keys.ENTER)
Brand=[]
Discription=[]
Price=[]
start=0
end=3
for page in range(start,end):
    Brands=driver.find_elements(By.XPATH,'//div[@class="_2WkVRV"]')
    discriptions=driver.find_elements(By.XPATH,'//a[@class="IRpwTa"]')
    prices=driver.find_elements(By.XPATH,'//div[@class="_30jeq3"]')
    for i in Brands:
        title=i.text
        Brand.append(title)
    for i in discriptions[0:40]:
        title=i.text
        Discription.append(title)
    for i in prices:
        title=i.text
        Price.append(title)
    next_button=driver.find_element(By.XPATH,'//a[@class="_1LKTO3"]')
    next_button.click
    time.sleep(3) 
Brand=Brand[0:100]
Discription=Discription[0:100]
Price=Price[0:100]
df=pd.DataFrame({'Brand':Brand,'Discription':Discription,'Prices':Price})
df
```

```python
driver=webdriver.Chrome()
driver.get('https://www.jagranjosh.com/general-knowledge/list-of-all-prime-ministers-of-india-1473165149-1')
primeministers=[]
dob=[]
term=[]
remarks=[]
data = []
table = driver.find_element(By.XPATH,'//table[@border="1"]/tbody')
rows = table.find_elements(By.TAG_NAME,'tr')
for row in rows:
    cols = row.find_elements(By.TAG_NAME,'td')
    if len(cols) == 5:
        name = cols[1].text
        born_dead = cols[2].text
        term_of_office = cols[3].text
        remarks = cols[4].text
        data.append([name, born_dead, term_of_office, remarks])
df = pd.DataFrame(data, columns=['Name', 'Born-Dead', 'Term of Office', 'Remarks'])
df
```
