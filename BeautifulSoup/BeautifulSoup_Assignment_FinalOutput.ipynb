```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = "https://en.wikipedia.org/wiki/Main_Page"
response = requests.get(url)
soup = BeautifulSoup(response.content, "html.parser")
header_tags = soup.find_all(["h1", "h2", "h3", "h4", "h5", "h6"])
header_texts = [tag.get_text() for tag in header_tags]
df = pd.DataFrame(header_texts, columns=["Header"])
print(df)
```

```python
url = "https://www.cnbc.com/world/?region=world"
response = requests.get(url)
soup = BeautifulSoup(response.content, "html.parser")
headlines = [headline.text.strip() for headline in soup.find_all('div', class_="Card-titleContainer")]
times = [time.text.strip() for time in soup.find_all('time')]
news_links = [a['href'] for a in soup.find_all('a', class_="Card-title")]
a = {'Headline': headlines,
     'Time': times,
     'News Link': news_links}
df = pd.DataFrame.from_dict(a, orient='index')
df = df.transpose()
print(df)
```

```python
url = "https://www.journals.elsevier.com/artificial-intelligence/most-downloaded-articles"
response = requests.get(url)
soup = BeautifulSoup(response.content, "html.parser")
articles_container = soup.find("ul",class_="sc-9zxyh7-0 cMKaMj")
titles = []
authors = []
dates = []
urls = []
for article in articles_container.find_all("li"):
  title = article.find("h2").text.strip()
  titles.append(title)
  author = article.find("span", class_="sc-1w3fpd7-0 dnCnAO").text.strip()
  authors.append(author)
  date = article.find("span", class_="sc-1thf9ly-2 dvggWt").text.strip()
  dates.append(date)
  url = article.find("a")["href"]
  urls.append(url)
data = {
  "Paper Title": titles,
  "Authors": authors,
  "Published Date": dates,
  "Paper URL": urls
}
df = pd.DataFrame(data)
print(df)
```

```python
url = "https://www.dineout.co.in/delhi-restaurants/buffet-special"
response = requests.get(url)
soup = BeautifulSoup(response.content, "html.parser")
restaurant_names = soup.find_all('a', class_='restnt-name ellipsis')
restaurant_list = []
for name in restaurant_names:
 restaurant_list.append(name.text.strip())
cuisines = soup.find_all('span', class_='double-line-ellipsis')
cuisine_list = []
for cuisine in cuisines:
  cuisine_list.append(cuisine.text.strip())
locations = soup.find_all('div', class_="restnt-loc ellipsis")
location_list = []
for location in locations:
  location_list.append(location.text.strip())
ratings = soup.find_all('div', class_="restnt-rating rating-4")
rating_list = []
for rating in ratings:
  rating_list.append(rating.text.strip())
image_urls = soup.find_all('img', class_='no-img')
image_url_list = []
for image in image_urls:
  image_url_list.append(image['data-src'])
data = {
  'Restaurant Name': restaurant_list,
  'Cuisine': cuisine_list,
  'Location': location_list,
  'Ratings': rating_list,
  'Image URL': image_url_list
}
df = pd.DataFrame(data)
print(df)
```
