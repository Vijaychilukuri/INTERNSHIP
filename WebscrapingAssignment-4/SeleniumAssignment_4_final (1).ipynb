```python
import selenium
import pandas as pd
import time
from bs4 import BeautifulSoup 
from selenium import webdriver
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException
import requests
import re
from selenium.webdriver.common.by import By
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.keys import Keys
import time
```

```python
url = "https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos"
driver=webdriver.Chrome()
driver.get(url)
time.sleep(5)
table = driver.find_element(By.XPATH,'/html/body/div[2]/div/div[3]/main/div[3]/div[3]/div[1]/table[2]/tbody')
rank_list = []
name_list = []
artist_list = []
upload_date_list = []
views_list = []
rows = table.find_elements(By.TAG_NAME, "tr")
for row in rows:
    columns = row.find_elements(By.TAG_NAME, "td")
    rank = columns[0].text.strip()
    name = columns[1].text.strip()
    artist = columns[2].text.strip()
    upload_date = columns[3].text.strip()
    views = columns[4].text.strip()

    rank_list.append(rank)
    name_list.append(name)
    artist_list.append(artist)
    upload_date_list.append(upload_date)
    views_list.append(views)

df=pd.DataFrame({'Song':rank_list,'Uploader':name_list,'Views_billions':artist_list,'Date':upload_date})
driver.quit()
```

```python
driver.get('https://archive.ics.uci.edu/datasets')
name=[]
description=[]
types=[]
instances=[]
features=[]
```

```python
na=driver.find_elements(By.XPATH,'//a[@class="link-hover link text-xl font-semibold"]')
for i in na:
    if i.text is None :
        name.append("--") 
    else:
        name.append(i.text)
de=driver.find_elements(By.XPATH,'//p[@class="truncate"]')
for i in de:
    if i.text is None :
        description.append("--") 
    else:
        description.append(i.text)
ty=driver.find_elements(By.XPATH,'//span[@class="truncate"][1]')
for i in ty:
    if i.text is None :
        types.append("--") 
    else:
        types.append(i.text)
clss=[]
for i in range(0,len(types),4):
    clss.append(types[i])
dtype=[]
for i in range(1,len(types),4):
    dtype.append(types[i])
instances=[]
for i in range(2,len(types),4):
    instances.append(types[i])
features=[]
for i in range(3,len(types),4):
    features.append(types[i])
DataSets=pd.DataFrame([])
DataSets['Dataset_Name']=name
DataSets['Description']=description
DataSets['Type']=clss
DataSets['DataType']=dtype
DataSets['Instances']=instances
DataSets['Features']=features
DataSets        
```
